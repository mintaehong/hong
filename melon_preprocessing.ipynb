```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 멜론 월간 차트 테이블
melon_chart = pd.read_csv(f'C:\\Users\\min\\melon_rank.csv')
melon_chart
```

```python
# 가수 테이블
singer_table = pd.read_csv(f'C:\\Users\\min\\singer_table.csv')

singer_table
```

```python
# 회사 테이블
company_table = pd.read_csv(f'C:\\Users\\min\\company_table.csv')

company_table
```

```python
# 테이블 병합
m_table = pd.merge(melon_chart, singer_table, how='left', on='가수')
m_table = pd.merge(m_table,company_table, how='left', on='소속사')
m_table.head()
```

```python
# 음악 장르 정규화
m_table.insert(4,'장르2',np.nan)
for genre in ['국내드라마', '국내영화', '인디음악']:
    s_table = m_table[m_table['장르'].str.contains(genre)]['장르'].str.split(', ')
    for v, i in zip(s_table, s_table.index):
        m_table.iloc[i,4] = v[v.index(genre)]
        del s_table[i][s_table[i].index(genre)]
        m_table.iloc[i,3]  = s_table[i][0]
print(m_table['장르'].unique())
print(m_table['장르2'].unique())
```

```python
# 활동유형 반점 제거
m_table['활동유형'] = m_table['활동유형'].str.replace(',', '')
```

```python
m_table
```

```python
# 멜론 월간 차트 탑100 차트인 많이 한 가수 순위
singer_c_table = pd.DataFrame(data = {'가수': m_table['가수'].value_counts().index, '개수': m_table['가수'].value_counts().values})
singer_c_table.index = singer_c_table.index + 1
singer_c_table.head(10)
```

```python
# 멜론 월간 차트 탑100 오래 유지한 노래 순위
name_c_table = pd.DataFrame(data = {'노래제목': m_table['제목'].value_counts().index, '개수': m_table['제목'].value_counts().values})
name_c_table.index = name_c_table.index + 1
name_c_table.head(10)
```

```python
# 멜론 월간 차트 탑100 차트인 많이 한 회사 순위 (상위 회사로 카운팅)

com_c_table = pd.DataFrame(data = {'회사': m_table['모회사'].value_counts().index, '개수': m_table['모회사'].value_counts().values})

# 가수 소속사 이중 소속 처리
for com, count in zip(com_c_table['회사'],com_c_table['개수']):
    if ',' in com:
        com_c_table.iloc[com_c_table[com_c_table['회사'] == com.split(',')[0]].index,1] += count
        com_c_table.iloc[com_c_table[com_c_table['회사'] == com.split(',')[1]].index,1] += count
        com_c_table.drop(index=com_c_table[com_c_table['회사'] == com].index, inplace = True)

com_c_table = com_c_table.sort_values('개수', ascending = False).reset_index(drop = True)
com_c_table.index=com_c_table.index + 1
com_c_table.head(10)
```

```python
# 멜론 월간 차트 탑100 장르 빈도 분포 확인

# 해당 장르의 곡 10개 이하는 기타 처리
frequency = {}
for v, i in zip(m_table['장르'].value_counts(), m_table['장르'].value_counts().index):
    if v >=10:
        frequency[i] = v
    else:
        try:
            frequency['기타'] += v
        except:
            frequency['기타'] = v

# 파이 그래프 사용
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rc('font', size=12)
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
colors = sns.color_palette('hls',len(frequency.keys()))
plt.figure(figsize=(10,10))
plt.pie(frequency.values(), labels = frequency.keys(), autopct='%.1f%%', wedgeprops = wedgeprops, colors=colors)
plt.title('장르')
plt.show()
```

```python
# 멜론 월간 차트 탑100 장르 빈도 분포 확인

# 활동 유형 솔로로 표기 되어있는 데이터 확인
# 확인 후 남성 솔로로 변경
m_table.iloc[m_table[m_table['활동유형'] == '솔로'].index,6] = '남성 솔로'

# 파이 그래프 사용
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
colors = sns.color_palette('hls',len(m_table['활동유형'].value_counts()))
plt.figure(figsize=(10,10))
plt.pie(m_table['활동유형'].value_counts(), labels = m_table['활동유형'].value_counts().index, autopct='%.1f%%', wedgeprops = wedgeprops, colors=colors)
plt.title('활동유형')
plt.show()
```
